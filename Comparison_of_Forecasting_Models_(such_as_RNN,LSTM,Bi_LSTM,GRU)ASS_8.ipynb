**1.Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU, Bidirectional
```

**2.Load & Preprocess Dataset**

```python
df = pd.read_csv("stock_data.csv")  # Use your dataset
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df = df[['Close']].dropna()
```

**3.EDA & Stationary Check**

```python
df['Close'].plot(title='Stock Prices Over Time')
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['Close'])
print(f'ADF Statistic: {result[0]}, p-value: {result[1]}')
```

**If non-stationary**

```python
df_diff = df.diff().dropna()
```

**4.Normalize and Prepare Data**

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

def create_sequences(data, seq_length=60):
    x, y = [], []
    for i in range(seq_length, len(data)):
        x.append(data[i - seq_length:i])
        y.append(data[i])
    return np.array(x), np.array(y)

sequence_length = 60
x, y = create_sequences(scaled_data, sequence_length)
split = int(0.8 * len(x))
x_train, x_test = x[:split], x[split:]
y_train, y_test = y[:split], y[split:]
```

**5.Model Implementations**

**RNN**

```python
rnn = Sequential()
rnn.add(SimpleRNN(50, return_sequences=False, input_shape=(x_train.shape[1], 1)))
rnn.add(Dense(1))
rnn.compile(optimizer='adam', loss='mse')
rnn.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

**LSTM**

```python
lstm = Sequential()
lstm.add(LSTM(50, return_sequences=False, input_shape=(x_train.shape[1], 1)))
lstm.add(Dense(1))
lstm.compile(optimizer='adam', loss='mse')
lstm.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

**Bi-LSTM**

```python
bi_lstm = Sequential()
bi_lstm.add(Bidirectional(LSTM(50), input_shape=(x_train.shape[1], 1)))
bi_lstm.add(Dense(1))
bi_lstm.compile(optimizer='adam', loss='mse')
bi_lstm.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

**GRU**

```python
gru = Sequential()
gru.add(GRU(50, return_sequences=False, input_shape=(x_train.shape[1], 1)))
gru.add(Dense(1))
gru.compile(optimizer='adam', loss='mse')
gru.fit(x_train, y_train, epochs=10, batch_size=32, verbose=1)
```

**6.Forecast Future Points**

```python
def forecast_future(model, last_sequence, n_steps=10):
    forecast = []
    current_seq = last_sequence.copy()

    for _ in range(n_steps):

        pred = model.predict(current_seq.reshape(1, *current_seq.shape))

        forecast.append(pred[0,0])
        current_seq = np.append(current_seq[1:], pred, axis=0)

    return np.array(forecast)
last_seq = x_test[-1]

# Forecast next 20 time points using LSTM (change model as needed)
future_steps = 20
forecast_scaled = forecast_future(lstm, last_seq, n_steps=future_steps)
forecast_actual = scaler.inverse_transform(forecast_scaled.reshape(-1,1))

print("Next 20 forecasted values:")
print(forecast_actual)
```

**Plotting Actual vs Forecasted Valus**

```python
import matplotlib.pyplot as plt

def get_predictions(model, X_test):
    preds_scaled = model.predict(X_test)
    preds = scaler.inverse_transform(preds_scaled)
    return preds

# Get predictions from each model
rnn_preds = get_predictions(rnn, x_test) # Changed X_test to x_test
lstm_preds = get_predictions(lstm, x_test) # Changed X_test to x_test
bi_lstm_preds = get_predictions(bi_lstm, x_test) # Changed X_test to x_test
gru_preds = get_predictions(gru, x_test) # Changed X_test to x_test

# Ground truth
y_true = scaler.inverse_transform(y_test)

# Plot all models together
plt.figure(figsize=(14, 6))
plt.plot(y_true[:100], label='Actual', color='black')  # Plot only first 100 for visibility
plt.plot(rnn_preds[:100], label='RNN', linestyle='--')
plt.plot(lstm_preds[:100], label='LSTM', linestyle='--')
plt.plot(bi_lstm_preds[:100], label='Bi-LSTM', linestyle='--')
plt.plot(gru_preds[:100], label='GRU', linestyle='--')

plt.title('Forecast Comparison: Actual vs Predicted (First 100 points)')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**7.Evaluation**

```python
results = {}
def evaluate_forecasts(actual, predicted, scaler):
    predicted_actual_scale = scaler.inverse_transform(predicted.reshape(-1, 1))
    actual_actual_scale = scaler.inverse_transform(actual.reshape(-1, 1))

    rmse = math.sqrt(mean_squared_error(actual_actual_scale, predicted_actual_scale))
    mae = mean_absolute_error(actual_actual_scale, predicted_actual_scale)
    mape = np.mean(np.abs((actual_actual_scale - predicted_actual_scale) / (actual_actual_scale + 1e-8))) * 100 # Added small epsilon for stability
    return rmse, mae, mape

for model, name in zip([rnn, lstm, bi_lstm, gru], ['RNN', 'LSTM', 'Bi-LSTM', 'GRU']):
    y_pred = model.predict(x_test)

    rmse, mae, mape = evaluate_forecasts(y_test, y_pred, scaler)
    results[name] = {'RMSE': rmse, 'MAE': mae, 'MAPE': mape}

# Display results in table form
import pandas as pd
results_df = pd.DataFrame(results).T
print(results_df)
```
