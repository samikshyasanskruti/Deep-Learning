```python
data=yf.download(ticker)
```

```python
#use close column analysis
df=data[['Close']]
```

```python
#plot the time series
plt.figure(figsize=(15,3))
plt.plot(df['Close'],label='Close price')
plt.title(f'(ticker)Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

**Fetching the data using a specified period**

```python
ticker='tsla'
period='1y'
data=yf.download(ticker,period=period)
data.tail()
df=data[['Close']]

#plot the time series
plt.figure(figsize=(15,3))
plt.plot(df['Close'],label='Close price')
plt.title(f'(ticker)Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

**Fetching the data using a specified period and interval**

```python
ticker='tsla'
period='1mo'
interval='5m'
data=yf.download(ticker,period=period,interval=interval)
data.tail()
df=data[['Close']]

#plot the time series
plt.figure(figsize=(15,3))
plt.plot(df['Close'],label='Close price')
plt.title(f'(ticker)Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```

**Fetching data using start and end date(YYYY-MM-DD)**

```python
ticker='tsla'
start='2021-04-01'
end='2025-03-31'

data=yf.download(ticker,start=start,end=end)
data.tail()
df=data[['Close']]

#plot the time series
plt.figure(figsize=(15,3))
plt.plot(df['Close'],label='Close price')
plt.title(f'(ticker)Stock Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid()
plt.show()
```
